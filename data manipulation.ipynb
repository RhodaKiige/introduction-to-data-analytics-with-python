##grouping and aggregation
we want to group data by department

```python
#group by syntax= df.groupby("condition")
grouped=df.groupby("department")
```

```python
grouped
```

```python
groupd=df.groupby("department")#apply functions ie calculations in order for python to create a table
print(groupd)
```

```python
df.groupby("department")["salary"].mean()
```

```python
df.groupby("department").agg({"salary":["sum", "mean","count","min","max","std","var","median"]})
```

```python
#give a summary of sum, mean, std, cout, max, min and median of all employees grouped by department and median 
df.groupby("department").agg({"salary":["sum", "mean", "count", "min","max","var","median"]})
```
